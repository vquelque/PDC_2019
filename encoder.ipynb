```python
import bitarray
import random
import string
import numpy as np
```

```python
N = 160
ba = bitarray.bitarray()
text = ''.join(random.choices(string.ascii_uppercase + string.ascii_lowercase + string.digits, k=N))
print(text)
ba.frombytes(text.encode('utf-8'))
array = np.array(ba.tolist()).astype(int)
print(ba)
result = ba.tobytes().decode('utf-8')
print(result == text)
print(array)
```
